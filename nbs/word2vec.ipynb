```python
#Goal: create word vectors from a GoT dataset
#analyze them to see word similarities
#Step0 - Import dependencies
from __future__ import absolute_import, division, print_function
#for word encoding
import codecs
#regex
import glob
#concurrency
import multiprocessing
#dealing with OS, like reading a file
import os
#pretty printing
import pprint
#regular expression
import re
#natural language toolkit
import nltk
#word 2 vec
import gensim.models.word2vec as w2v
#dimensionality reduction
import sklearn.manifold
#math
import numpy as np
#plotting
import matplotlib.pyplot as plt
#parse
import pandas as pd
#visualization
import seaborn as sns
```

```python
#Step 1 process our data
#clean data
nltk.download('punkt') #pretrained tokenizer
nltk.download('stopwords') #words like and, the, an, a, of
```

```python
#get the book names, matching txt files
book_filenames = sorted(glob.glob("data/got/*.txt"))
book_filenames
```

```python
corpus_raw = u""
for book_filename in book_filenames:
    print("Reading '{0}'...".format(book_filename))
    with codecs.open(book_filename, "r", "utf-8") as book_file:
        corpus_raw += book_file.read()
    print(f"Corpus is now {len(corpus_raw)} characters long")
    print()
```

```python
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
```

```python
raw_sentences = tokenizer.tokenize(corpus_raw)
```

```python
def sentence_to_wordlist(raw):
    clean = re.sub("[^a-zA-Z]", " ", raw)
    words = clean.split()
    return words
```

```python
sentences = []
for raw_sentence in raw_sentences:
    if len(raw_sentence) > 0:
        sentences.append(sentence_to_wordlist(raw_sentence))
```

```python
raw_sentences[10]
sentence_to_wordlist(raw_sentences[10])
```

## Train Word2Vec

```python
#ONCE we have vectors

#3 main tasks that vectors help with
# DISTANCE, SIMILARITY, RANKING

#dimensionality of the resulting word vectors.
#more dimensions, more computationally expensive to train
#but also more accurate
#more dimensions = more generalized
num_features = 300

#minimum word count threshold
min_word_count = 3

#number of threads to run in parallel
#more workers, faster we train
num_workers = multiprocessing.cpu_count()

#context window length
context_size = 7

#downsample setting for frequent words.
#0 - 1e-5 is good for this
downsampling = 1e-3

#seed for the rng, to make results repruducible
#deterministic, good for debugging
seed = 1
```

```python
thrones2vec = w2v.Word2Vec(
    sg=1,
    seed=seed,
    workers=num_workers,
    size=num_features,
    min_count=min_word_count,
    window=context_size,
    sample=downsampling
)
```

```python
thrones2vec.build_vocab(sentences)
```

```python
thrones2vec.train(sentences, total_examples=len(sentences), epochs=1)
```

## Compress the word vectors into 2D space and plot them

```python
#WATCH VIDEO - how to visualize a dataset easily
tsne = sklearn.manifold.TSNE(n_components=2, random_state=0)
```

```python
all_word_vectors_matrix = thrones2vec.trainables.syn1neg
all_word_vectors_matrix
```

```python
#train t sne
all_word_vectors_matrix_2d = tsne.fit_transform(all_word_vectors_matrix)
```

```python
#plot point in 2d space
points = pd.DataFrame(
    [
        (word, coords[0], coords[1])
        for word, coords in [
            (word, all_word_vectors_matrix_2d[thrones2vec.wv.vocab[word].index])
            for word in thrones2vec.wv.vocab
        ]
    ],
    columns=["word", "x", "y"]
)
```

```python
def plot_region(x_bounds, y_bounds):
    slice = points[
        (x_bounds[0] <= points.x) &
        (points.x <= x_bounds[1]) & 
        (y_bounds[0] <= points.y) &
        (points.y <= y_bounds[1])
    ]
    
    ax = slice.plot.scatter("x", "y", s=35, figsize=(10, 8))
    for i, point in slice.iterrows():
        ax.text(point.x + 0.005, point.y + 0.005, point.word, fontsize=11)
```

```python
#distance, similarity, and ranking
def nearest_similarity_cosmul(start1, end1, end2):
    similarities = thrones2vec.wv.most_similar_cosmul(
        positive=[end2, start1],
        negative=[end1]
    )
    start2 = similarities[0][0]
    print("{start1} is related to {end1}, as {start2} is related to {end2}".format(**locals()))
    return start2
```
